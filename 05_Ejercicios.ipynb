# **Ejercicios**
---

1. **Operadores Aritméticos**
   - Dadas las variables `num1 = 10` y `num2 = 3`, calcula la suma de ambos y almacena el resultado en la variable `resultado_suma`.

2. **Expresiones Aritméticas**
   - Dadas las variables `a = 5`, `b = 2` y `c = 8`, calcula el resultado de la expresión `(a + b) * c` y almacena el resultado en la variable `resultado_expresion`.

3. **Descuento en una Compra**
   - Dado el monto de una compra `monto = 150` y un descuento del 20%, calcula el monto final con descuento y almacénalo en la variable `monto_con_descuento`.

4. **Operadores Relacionales**
   - Dadas las variables `x = 15` y `y = 7`, verifica si `x` es mayor que `y` y almacena el resultado en la variable `es_mayor`.

5. **Longitud de Cadena y Operador Relacional**
   - Calcula la longitud de la cadena "Python" y verifica si es igual a 6. Almacena el resultado en la variable `es_igual_longitud`.

6. **Par o Impar**
   - Dado el número `numero = 25`, verifica si es un número par y almacena el resultado en la variable `es_par`.

7. **Operador Identidad**
   - Dadas las cadenas `cadena1 = "Hola"` y `cadena2 = "Hola"`, verifica si ambas variables referencian al mismo objeto y almacena el resultado en la variable `son_iguales`.

8. **Calificación Aprobatoria**
   - Dada una calificación `calificacion = 85`, verifica si es igual o mayor que 70. Almacena el resultado en la variable `aprobado`.

9. **Operador Lógico AND**
   - Dadas las variables `valor1 = True` y `valor2 = False`, aplica el operador lógico AND y almacena el resultado en la variable `resultado_and`.

10. **Operador Lógico OR**
    - Dados los números `numero1 = 8` y `numero2 = 3`, verifica si `numero1` es menor que 5 o `numero2` es mayor que 2. Almacena el resultado en la variable `resultado_or`.

11. **Temperatura Calurosa**
    - Dada una temperatura `temperatura = 25`, verifica si es mayor que 30 o menor que 10. Almacena el resultado en la variable `es_caluroso`.

12. **Operador Lógico NOT**
    - Dada una variable de verdad `valor_verdad = True`, aplica el operador lógico NOT y almacena el resultado.

13. **División Entera**
    - Dados los números `dividendo = 15` y `divisor = 4`, realiza la división entera y almacena el cociente en la variable `cociente`.

14. **Verificación de Saldo**
    - Dado un saldo `saldo = 100` y un monto de retiro `retiro = 120`, verifica si se puede realizar el retiro sin exceder el saldo. Almacena el resultado en la variable `puede_retirar`.

15. **Búsqueda de Subcadena**
    - Dada la cadena `cadena3 = "Python"`, verifica si la subcadena "th" está presente en la cadena. Almacena el resultado en la variable `es_subcadena`.

16. **Divisibilidad**
    - Dados los números `numero3 = 12` y `numero4 = 5`, verifica si `numero3` es divisible por `numero4`. Almacena el resultado en la variable `es_divisible`.

17. **Diferencia y Operador Lógico NOT**
    - Dadas las variables `diferencia = 7 - 3`, verifica si la diferencia es diferente de 4. Almacena el resultado en la variable `es_diferente`.

18. **Comparación de Identidad**
    - Dadas las variables `numero5 = 6` y `numero6 = 6`, verifica si ambas variables son distintas objetos. Almacena el resultado en la variable `son_distintos`.

19. **Operador Lógico XOR**
    - Dadas las variables `variable1 = True` y `variable2 = False`, aplica el operador lógico XOR y almacena el resultado en la variable `resultado_xor`.





# **Soluciones**
---

```python
# Soluciones

# Ejercicio 1
num1 = 10
num2 = 3
resultado_suma = num1 + num2

# Ejercicio 2
a = 5
b = 2
c = 8
resultado_expresion = (a + b) * c

# Ejercicio 3
monto = 150
descuento = 0.2
monto_con_descuento = monto - (monto * descuento)

# Ejercicio 4
x = 15
y = 7
es_mayor = x > y

# Ejercicio 5
longitud_cadena = len("Python")
es_igual_longitud = longitud_cadena == 6

# Ejercicio 6
numero = 25
es_par = numero % 2 == 0

# Ejercicio 7
cadena1 = "Hola"
cadena2 = "Hola"
son_iguales = cadena1 is cadena2

# Ejercicio 8
calificacion = 85
aprobado = calificacion >= 70

# Ejercicio 9
valor1 = True
valor2 = False
resultado_and = valor1 and valor2

# Ejercicio 10
numero1 = 8
numero2 = 3
resultado_or = numero1 < 5 or numero2 > 2

# Ejercicio 11
temperatura = 25
es_caluroso = temperatura > 30 or temperatura < 10

# Ejercicio 12
valor_verdad = not True

# Ejercicio 13
dividendo = 15
divisor = 4
cociente = dividendo // divisor

# Ejercicio 14
saldo = 100
retiro = 120
puede_retirar = saldo >= retiro

# Ejercicio 15
cadena3 = "Python"
es_subcadena = "th" in cadena3

# Ejercicio 16
numero3 = 12
numero4 = 5
es_divisible = numero3 % numero4 == 0

# Ejercicio 17
diferencia = 7 - 3
es_diferente = not diferencia == 4

# Ejercicio 18
numero5 = 6
numero6 = 6
son_distintos = not numero5 is numero6

# Ejercicio 19
variable1 = True
variable2 = False
resultado_xor = variable1 != variable2


```
