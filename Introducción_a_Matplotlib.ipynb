## Introducción a Matplotlib

**Matplotlib** es una biblioteca de visualización de Python que proporciona diversas herramientas y funciones para crear gráficos y visualizaciones estáticas, animadas e interactivas. Es una de las bibliotecas más populares y utilizadas en la comunidad Python.

`pyplot` es un módulo de la biblioteca Matplotlib que proporciona una interfaz sencilla e intuitiva para crear gráficos. Suele ser el módulo utilizado por los ingenieros de Aprendizaje Automático y ciencia de datos para sus representaciones gráficas. En concreto, los puntos clave de este módulo son:

- **Interfaz de alto nivel**: pyplot ofrece una serie de funcionalidades que facilitan la creación rápida de gráficos.
- **Funcionalidad**: Ofrece una amplia variedad de funciones para gráficos de barras, de puntos, de cajas, etc.
- **Integración**: Está estrechamente integrado con entornos como Jupyter Notebook, lo que permite mostrar gráficos directamente dentro de los cuadernos.

```python
import numpy as np

X = np.linspace(0, 10, 100)
y = np.sin(X)
z = np.cos(X)
```

### Gráfico de líneas

El **gráfico de líneas** representa información en puntos conectados por líneas. Es útil para mostrar la evolución de una o varias series de datos a lo largo de un eje, normalmente el tiempo,

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10, 5))

plt.plot(X, y, label = "Sen X")
plt.plot(X, z, label = "Cos X")

plt.title("Gráfico de Líneas")
plt.legend()
plt.show()
```

### Gráfico de dispersión

El **trazado de dispersión** muestra los valores individuales de dos variables numéricas en un plano cartesiano (con dos ejes). Cada punto representa una observación.

```python
plt.figure(figsize = (10, 5))

plt.scatter(X, y, label = "Sen X")

plt.title("Gráfico de Dispersión")
plt.legend()
plt.show()
```

### Histograma

El **histograma** representa la distribución de una variable numérica dividiendo el rango de datos en intervalos y mostrando cuántos datos caen en cada intervalo (para variables continuas) o la frecuencia de cada categoría (para variables categóricas).

```python
data = np.random.randn(1000)

plt.figure(figsize = (10, 5))

plt.hist(data, bins = 30, alpha = 0.7)

plt.title("Histograma")
plt.show()
```

### Gráfico de barras

El **diagrama de barras** representa datos categóricos con barras rectangulares con alturas (o longitudes, en el caso de barras horizontales) proporcionales a los valores que representan.

```python
labels = ["A", "B", "C", "D"]
values = [10, 20, 15, 30]

plt.figure(figsize = (10, 5))

plt.bar(labels, values)

plt.title("Diagrama de Barras chart")
plt.show()
```

### Gráfico circular

Un **gráfico circular** representa los datos en sectores circulares, donde cada sector corresponde a una categoría y su tamaño es proporcional al valor que representa.

```python
labels = ["A", "B", "C", "D"]
sizes = [215, 130, 245, 210]

plt.figure(figsize = (7, 7))

plt.pie(sizes, labels = labels)

plt.title("Gráfico circular")
plt.show()
```

### Boxplot

Un **boxplot** muestra la distribución de datos cuantitativos por sus cuartiles y, posiblemente, los valores atípicos.

![Boxplot example](https://github.com/4GeeksAcademy/machine-learning-prework/blob/main/04-matplotlib/assets/boxplot.png?raw=true "boxplot")

Los extremos de la caja indican los cuartiles inferior y superior, mientras que la línea del interior de la caja indica la mediana.

```python
data = np.random.randn(1000)

plt.figure(figsize = (10, 5))

plt.boxplot(data)

plt.title("Boxplot")
plt.show()
```

#### Ejercicio 03: Crear un histograma a partir de una matriz aleatoria siguiendo una distribución normal $N ~ (2, 1.5) $ (🐸🐸🐸)

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.normal(2, 1.5, 1000)

plt.figure(figsize = (10, 5))
plt.hist(data, bins = 30, alpha = 0.7)
plt.title("Histograma")
plt.show()
```

#### Ejercicio 04: Crea un DataFrame a partir del conjunto de datos del Titanic y muestra las distribuciones de edad y coste de los billetes (🐸🐸🐸🐸)
